```python
#big o notation
def sum1(n):
    final_sum= 0 
    
    for x in range(n+1):
        final_sum += x
    return final_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return (n*(n+1)/2)
```

```python
sum2(10)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Set up runtime comparisons
n = np.linspace(1,10,1000)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot setup
plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```

```python
#O(1) Constant
def func_constant(values):
    print(values[0])
    
```

```python
lst = [1, 2, 3, 4, 5 , 6]
func_constant(lst)
```

```python
# O(n) Linear - print statement prints linearly with n
def func_lin(lst):
    for val in lst:
        print(val)
```

```python
func_lin(lst)
```

```python
# O(n^2) Quadratic
def func_qad(lst):
    for it1 in lst:
        for it2 in lst:
            print (it1, it2)
```

```python
func_qad(lst)
```

```python
def print_once(lst): #O(n)
    for val in lst:
        print(val)
print_once(lst)        
```

```python
def print_two(lst): #O(n)
    for val in lst:
        print(val)
    for val in lst:
        print(val)
```

```python
print_two(lst)
```

```python
def comp(lst):
    print (lst[0]) #O(n)
    
    ### O(n/2) .... O(1/2 * n)
    mp=len(lst)//2
    for val in lst[:mp]:
        print(val)
        
    ### O(10)
    for x in range(10):
        print ("hello world")
comp(lst)
```

```python
def matcher(lst, match):
    for item in lst:
        if item == match:
            return True
    return False
```

```python
def create_list(n):
    new_lst = []
    
    for num in range(n):
        new_lst.append('new')
    return new_lst
```
